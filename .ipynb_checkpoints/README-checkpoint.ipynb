# Advanced Lane Finding Project
The goals / steps of this project are the following:
* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

## Camera Calibration

* I've started by calibrating the camera using the images supplied in the "camera_cal" folder.
* This was done using the cv2 library functions findChessboardCorners and calibrateCamera that ultimately gave me the camera distortion coefficients. After this, everytime I wanted to undistort an image taken from that camera I used the cv2.undistort() function.
* Here are 2 examples of "camera undistortion":

```python
import matplotlib.pyplot as plt
image = plt.imread('./output_images/cameracal.png')
plt.imshow(image)
plt.show()

image = plt.imread('./output_images/undistortroad.png')
plt.imshow(image)
plt.show()
print("After comparing the two images, I believe that the image was correctly undistorted by seeing that the traffic sign is different and the bonnet looks more even.")
```

## Image processing
* (SUBMISSION 1): After undistort the images I've applied to each image sobelx and gradient magnitude combined with HLS (S channel and H channel) color thresholds.

* (SUBMISSION 2): Following the review 1 suggestions I've decided to experiment with different color spaces and thresholds to find a combination which enabled me to accurately detect the lane pixels withouth getting distracted by shadows and color/brightness changes. Therefore, for the submission I've used a combination of LUV anc LAB color spaces. There is less noise and the lanes are more clear now.

* Here is a before and after example:

```python
image = plt.imread('./output_images/exampleofbinary2.png')
plt.imshow(image)
plt.show()
```

## Perspective transform
* Now that I have the binary image I've applied perspective transform to the image.
* First I confined the lane lines into a trapezoidal shape and then I brought this new region into a bird's eye view using cv2.warpPerspective and cv2.getPerspectiveTransform.

```python
image = plt.imread('./output_images/regionmask.png')
plt.imshow(image)
plt.show()

image = plt.imread('./output_images/birdseye.png')
plt.imshow(image)
plt.show()

print("applying binary and \" bird's eye view together to other image I get...")
image = plt.imread('./output_images/birdeyeplusbinarycurve.png')
plt.imshow(image)
plt.show()
```

## Locating the lines

* After this, I've used the lectures pipeline to locate the lane lines.
* Starting by getting the processed image's histogram and identifying the most "obvious/clear" lines where we try to describe them by fitting a 2nd degree polynomial function.
* (SUBMISSION 2): To prevent picking up noise from the motorway hardshoulder the histogram ignored the first and last 200 ("horizontal axis") pixels of the image

```python
image = plt.imread('./output_images/histogram.png')
plt.imshow(image)
plt.show()
print("can clearly see 2 peaks on this pixel summing histogram, meaning that are 2 clear lines on this image")
```

```python
image = plt.imread('./output_images/locatinglines.png')
plt.imshow(image)
plt.show()

image = plt.imread('./output_images/fittingapoly.png')
plt.imshow(image)
plt.show()

print("yes, I can find a 2nd order polynomial function that can somehow represent each of the lines")
print("if I can fit a function, I can take it's curvature in relation to any axis...")
```
